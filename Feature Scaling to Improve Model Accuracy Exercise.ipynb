#Getting X and y Separated

```python
y = df['Purchased']
```

```python
X = df[['Age', 'Salary']]
```

#Getting Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, stratify=y, random_state = 2529)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

#Getting Scaling of Features or Attributes or Independent Variable X

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
X_train_ss = ss.fit_transform(X_train)
```

```python
X_test_ss = ss.fit_transform(X_test)
```

#Getting Visualization of Impact of Scaling 

```python
X_train_ss = pd.DataFrame(X_train_ss, columns = X_train.columns)
X_test_ss = pd.DataFrame(X_test_ss, columns = X_test.columns)
```

#Reduced Impact of Outlier

```python
fig, (ax1,ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.scatter(X_train['Age'], X_train['Salary'])
ax1.set_title('Before Scaling')
ax2.scatter(X_train_ss['Age'], X_train_ss['Salary'], color = 'red')
ax2.set_title('After Scaling')
plt.show()
```

#Uniform Scale and Distribution

```python
fig, (ax1,ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.set_title('Before Scaling')
sns.kdeplot(X_train["Age"], ax=ax1)
sns.kdeplot(X_train["Salary"], ax=ax1)

ax2.set_title('After Scaling')
sns.kdeplot(X_train_ss["Age"], ax=ax2)
sns.kdeplot(X_train_ss["Salary"], ax=ax2)
plt.show()
```

#Comparing Distrubution Before Scaling and After Scaling

```python
fig, (ax1,ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.set_title('Age Distribution Before Scaling')
sns.kdeplot(X_train["Age"], ax=ax1)

ax2.set_title('Age Distribution After Scaling')
sns.kdeplot(X_train_ss["Age"], ax=ax2)
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.set_title('Salary Distribution Before Scaling')
sns.kdeplot(X_train["Salary"], ax=ax1)

ax2.set_title('Salary Distribution After Scaling')
sns.kdeplot(X_train_ss["Salary"], ax=ax2)
plt.show()
```

#Logistic Regression one by one on both Raw X and Scale X

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```

```python
lr.fit(X_train_ss, y_train)
```

```python
y_pred = lr.predict(X_test_ss)
```

```python
accuracy_score(y_test, y_pred)
```
