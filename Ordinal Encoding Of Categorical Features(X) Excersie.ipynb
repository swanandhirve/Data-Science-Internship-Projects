#Getting Features (X) Separated from DataFrame 

```python
X=df[['Gender', 'Education', 'Review']]
```

```python
X
```

#Encoding Categorical Features As An Array

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder()
```

```python
X = oe.fit_transform(X)
```

```python
X
```

#Displaying Encoded Categories

```python
oe.categories_
```

#Reverse Encoding Integer Array as Categorical Features (X) Label

```python
oe.inverse_transform([[0, 0, 0]])
```

```python
oe.inverse_transform([[1, 1, 1]])
```

```python
oe.inverse_transform([[1, 2, 2]])
```
